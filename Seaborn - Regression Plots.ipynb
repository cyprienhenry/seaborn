```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('./data/college_datav3.csv')
```

```python
df.head()
```

## regplot()

```python
sns.regplot(data=df,
           x='SAT_AVG_ALL',
           y='Tuition', 
           marker='.')

plt.show()
```

## regplot() with estimator for x

```python
sns.regplot(data=df,
           x='REGION',
           y='Tuition',
           x_estimator=np.mean,
           order=2)

plt.show()
```

## residplot()

```python
sns.residplot(data=df,
           x='SAT_AVG_ALL',
           y='Tuition')

plt.show()
```
