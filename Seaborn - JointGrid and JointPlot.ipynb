```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/bike_share.csv')
```

# JointGrid and jointplot

```python
sns.set_style('whitegrid')
g = sns.JointGrid(x='hum', y='total_rentals', data=df, xlim=(0.1, 1.0))
g.plot(sns.regplot, sns.distplot)
```

```python
sns.jointplot(x="temp",
         y="total_rentals",
         kind='reg',
         data=df,
         order=2,
         xlim=(0, 1))
```

# Visualize the residuals in a regression

```python
sns.jointplot(x="temp",
        y="total_rentals",
        kind='residplot',
        data=df,
        order=2)
```

## Overlaying plots

```python
g = (sns.jointplot(x="temp",
             y="casual",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
```
