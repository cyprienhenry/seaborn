```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/insurance_premiums.csv')
```

```python
df.head()
```

# pairgrid

```python
g = sns.PairGrid(data=df, vars=['fatal_collisions', 'premiums'])
g.map(plt.scatter)
```

```python
g = sns.PairGrid(df, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(sns.distplot)
g3 = g2.map_offdiag(plt.scatter)
```

## pairplot()

```python
sns.pairplot(data=df,
        vars=["fatal_collisions", "premiums"],
        kind='scatter',
        hue='Region',
        palette='RdBu',
        diag_kws={'alpha':.5})
```

```python
sns.pairplot(data=df,
             vars=["insurance_losses", "premiums"],
             kind='reg',
             palette='BrBG',
             diag_kind = 'kde',
             hue='Region')
```
